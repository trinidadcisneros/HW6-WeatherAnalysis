## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
cities
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Next, spend the requisite time necessary to study the OpenWeatherMap API. 
# Based on your initial study, you should be able to answer basic questions about the API: 
# Where do you request the API key? 
# Which Weather API in particular will you need? 
# api.openweathermap.org/data/2.5/weather?q={city name}
# api.openweathermap.org/data/2.5/weather?q={nizhneyansk}
# What URL endpoints does it expect? 
# What JSON structure does it respond with? 
# Before you write a line of code, 
# you should be aiming to have a crystal clear understanding of your intended outcome.

url = "http://api.openweathermap.org/data/2.5/weather?"
city_data = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []

# Get weather data
print("Beginning Data Retrieval")
print("------------------------------")

index = 1
for city in cities:
    query_url = f"{url}appid={api_key}&q={city}"
    weather_response = requests.get(query_url)
    weather_json = weather_response.json()
    try:
        lat.append(weather_json["coord"]["lat"])      
    except: pass
    try:
        lng.append(weather_json["coord"]["lon"])
    except: pass
    try:
        humidity.append(weather_json["main"]["humidity"])
    except: pass
    try:
        max_temp.append(weather_json["main"]["temp_max"])
    except: pass
    try:
        wind_speed.append(weather_json["wind"]["speed"])
    except: pass
    try:
        cloudiness.append(weather_json["clouds"]["all"])
    except: pass
    try:
        country.append(weather_json["sys"]["country"])
    except: pass        
    try:
        date.append(weather_json['dt'])
    except: pass
    try:
        city_data.append(weather_json["name"])
        print(f"Processing Record {index} of Set {index} | {city}") 
    except (NameError, KeyError):
        print("Not found. Skipping...")    
    index = index + 1      
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.

```python
data = [city, cloudiness, country, date, humidity, lat, lng, max_temp, wind_speed]
col_names = ["City", "Cloudiness", "Country", "Date", "Humidity", "Lat", "Lng", "Max Temp", "Wind Speed"]

weather_dict = {
    "City": city_data,
    "Cloudiness": cloudiness,
    "Country": country,
    "Date" : date,
    "Humidity" : humidity,
    "Lat": lat,
    "Lng": lng,
    "Max Temp" : max_temp,
    "Wind Speed" : wind_speed
}

df = pd.DataFrame(data = weather_dict)
df
```
